```python
# load library
import pandas as pd
# create dataframe
dataframe = pd.DataFrame()
# add columns
dataframe['Name'] = ['Lemon','Apple','Banana']
dataframe['Price'] = [5,8,7]
dataframe['License'] = [False,True,True]
# show data
dataframe
```

```python
dataframe.shape
```

```python
# df la dataframe
df = dataframe
df.info()
```

```python
# print columns name
list(df.columns)
```

```python
df.describe()
# count:so luong
# mean:gia tri trung binh
# std:do lech chuan
```

```python
# loc iloc (location & index location)
df.loc[(df.Price == 5)]
```

```python
df.loc[(df.Price == 3) | (df.Name == "Banana")]
```

```python
df.loc[(df.Price == 5) & (df.Name == "Lemon")]
```

```python
df.loc[(df.Price == 5)| (df.Name == 'Banana')]
```
